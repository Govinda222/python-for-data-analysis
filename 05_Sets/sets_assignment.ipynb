### Problem 3: Set Operations

Create two sets: one with the first 5 positive integers and another with the first 5 even integers. Perform and print the results of union, intersection, difference, and symmetric difference operations on these sets.

```python
positive_set = {number for number in range(1,6)}
even_set = {number for number in range (1,11) if number%2 == 0 }
print(f"First five positive number set is {positive_set}")
print(f"first five even number set is {even_set}")
union_set = positive_set.union(even_set)
intersection_set = positive_set.intersection(even_set)
difference_set = positive_set.difference(even_set)
symmetric_difference_set = positive_set.symmetric_difference(even_set)
print(f"The union of even and positive set is {union_set}")
print(f"The intersection of even and positive set is {intersection_set}")
print(f"The difference of even and positive set is {difference_set}")
print(f"The symentic_difference of even and positive set is {symmetric_difference_set}")
```

### Problem 4: Set Comprehensions

Create a new set containing the squares of the first 10 positive integers using a set comprehension. Print the new set.

### Problem 5: Filtering Sets

### Problem 6: Set Methods

Create a set with duplicate elements and remove the duplicates using set methods. Print the modified set.

```python
import random
duplicate_list = [random.randint(1,10) for _ in range (100)] # making a duplicate list instead as set doesnot allow duplicates 
print(f"The duplicate list is {duplicate_list}")
non_duplicate_filtered_set = set(duplicate_list)
print(f"The filtered list is {non_duplicate_filtered_set}")
```

### Problem 7: Subsets and Supersets

Create two sets: one with the first 5 positive integers and another with the first 3 positive integers. Check if the second set is a subset of the first set and if the first set is a superset of the second set. Print the results.

```python
subset_set = {number for number in range (1,4)}
superset_set = {number for number in range (1,6)}
print(f"checking is the subset actually the subset {subset_set.issubset(superset_set)}")
print(f"checking is the superset actually the superset {superset_set.issuperset(subset_set)}")
```

### Problem 8: Frozenset

### Assignment 9: Set and List Conversion

Create a set with the first 5 positive integers. Convert it to a list, append the number 6, and convert it back to a set. Print the resulting set.

```python
numbers2 = {number for number in range (1,6)}
numbers2 = list(numbers2)
print(numbers2)
print(type(numbers2))
numbers2.append(6)
numbers2 = set(numbers2)
print(numbers2)
```
