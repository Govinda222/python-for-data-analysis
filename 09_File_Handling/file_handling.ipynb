```python
#opening a file 
def open_file(address):
    with open(address) as text_file:
        file = text_file.read()
    return file
```

```python
# writing a file
def write_file (address, message):
    with open(address ,mode="w") as text_file :
        text_file.write(message)
```

```python
#appending a file 
def append_file (address,message): 
    with open(address, mode="a") as text_file : 
        text_file.write(message)
```

```python
#emptying a file
def clear_file (address):
    with open(address, mode="w") as text_file:
        text_file.write("")
```

```python
# Counting lines, words, and characters in a text file
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
lines, words, characters = count_text_file(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
```

```python
# Copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
#### Using the os module
import os
cwd=os.getcwd()
print(f"Current working directory is {cwd}")
```

```python
## create a new directory
new_directory="package"
os.mkdir(new_directory)
print(f"Directory '{new_directory}' create")
```

```python
## Listing Files And Directories
items=os.listdir('.')
print(items)
```

```python
#Checking if a Path is a File or Directory
import os

path = 'example.txt'
if os.path.isfile(path):
    print(f"The path '{path}' is a file.")
elif os.path.isdir(path):
    print(f"The path '{path}' is a directory.")
else:
    print(f"The path '{path}' is neither a file nor a directory.")
```

```python
## Getting the absolute path
relative_path='example.txt'
absolute_path=os.path.abspath(relative_path)
print(absolute_path)
```
