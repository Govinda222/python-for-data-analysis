```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [3,56,73,78,90]
y = [23,45,75,99,245]
```

```python
plt.plot(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("Basic Plot")
plt.show()
```

```python
#customized line plot 
plt.plot(x,y,color = "red", linestyle="--",linewidth = 4, marker = ".",markersize= 20)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("Basic Plot")
plt.show()
```

```python
#Multiple Plots 
x = [i for i in range(1,6)]
print(x)
y = [i*i for i in range (1,6)]
print(y)
y2 = [i for i in range (1,6)]
print(y2)
x2 = [i*i for i in range(1,6)]
print(x2)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.plot(x,y,color="red")
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x2,y,color = "blue")
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color = "green")
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x2,y2,color = "yellow")
plt.title("Plot 4")

```

```python
df = pd.read_csv("data.csv")
display(df)
```

```python
# lets do some plotting using the data such as total sales of each product in pie chart and all
per_product_sales_data = (
    df.groupby("Product")["Sales"]
      .sum()
      .reset_index(name="Total_Sales")
)

display(per_product_sales_data)
```

```python
colours = ["gold","yellowgreen","lightcoral"]
plt.pie(per_product_sales_data["Total_Sales"],labels=per_product_sales_data["Product"],autopct="%1.2f%%",colors=colours,shadow=True)
```

```python
per_product_sales_data.plot(kind = 'bar',color="red")
```

```python
each_product_sales_per_region = df.groupby(["Product","Region"])["Sales"].sum()
display(each_product_sales_per_region)
```
