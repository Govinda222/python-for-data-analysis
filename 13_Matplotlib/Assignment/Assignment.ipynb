### Q20 ⭐ Challenge

Create **one dashboard** (using `plt.subplot()`) containing:

* 📊 Bar Chart → Sales by Region
* 🥧 Pie Chart → Category Distribution
* 📈 Line Plot → Sales over Time
* 📉 Histogram → Sales Distribution

````python
#16
pivot_sales = pd.pivot_table(
    df, 
    values="Sales", 
    index="Region", 
    columns="Category", 
    aggfunc="sum",
    fill_value=0
)
print("\n[Q16] Pivot Table (Sales by Region & Category):")
print(pivot_sales)
````

````python
#17
avg_sales_region = df.groupby("Region")["Sales"].mean().reset_index(name="Avg_Sales")
print("\nAverage Sales by Region:")
print(avg_sales_region)
````

````python
top_product = df["Product"].value_counts().idxmax()
print(f"\nMost Frequent Product: {top_product}")
````

````python
#19
best_day_idx = df["Sales"].idxmax()
best_day = df.loc[best_day_idx, "Date"]
best_day
````

````python
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Data Analyst Challenge Dashboard", fontsize=16, fontweight="bold")

# 1. Bar Chart -> Sales by Region
region_sales = df.groupby("Region")["Sales"].sum()
axs[0, 0].bar(region_sales.index, region_sales.values, edgecolor="black")
axs[0, 0].set_title("Total Sales by Region")
axs[0, 0].set_ylabel("Sales")

# 2. Pie Chart -> Category Distribution
cat_counts = df["Category"].value_counts()
axs[0, 1].pie(cat_counts, labels=cat_counts.index, autopct="%1.1f%%", startangle=140)
axs[0, 1].set_title("Category Distribution (Count)")

# 3. Line Plot -> Sales over Time
time_sales = df.groupby("Date")["Sales"].sum()
axs[1, 0].plot(time_sales.index, time_sales.values, color="teal", linewidth=2)
axs[1, 0].set_title("Sales Over Time")
axs[1, 0].set_ylabel("Sales")
axs[1, 0].tick_params(axis='x', rotation=45)

# 4. Histogram -> Sales Distribution
axs[1, 1].hist(df["Sales"], bins=20, color="purple", edgecolor="black", alpha=0.7)
axs[1, 1].set_title("Distribution of Sales")
axs[1, 1].set_xlabel("Sales Amount")
axs[1, 1].set_ylabel("Frequency")

plt.tight_layout()
````
