# 📦 Modules and Packages in Python

## Objective

Understand how Python organizes reusable code using modules, packages, and the standard library.

### Topics

- Modules
- Import Statements
- Standard Library
- Packages
- __init__.py

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4])
```

```python
from math import *
print(sqrt(16))
print(pi)
```

```python
from package.maths import addition
addition(2,3)
```

```python
from package import maths
maths.addition(2,3)
```
