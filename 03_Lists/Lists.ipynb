### Problem 8: List of Dictionaries

Create a list of dictionaries where each dictionary represents a student with keys 'name' and 'score'. Sort the list of dictionaries by the 'score' in descending order and print the sorted list.

```python
import random
names = ["Govinda","Anas","Ram","Aman","Priya"]
students = [{'name': name, 'score': random.randint(1,100)} for name in names]
print(students)
students.sort(key=lambda x: x['score'], reverse=True)
print(students)
```

### Problem 9: Matrix Transposition

Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.

```python
def get_transpose(matrix):
    transposed_matrix = [[matrix[j][i] for j in range(3)] for i in range(3)]
    return transposed_matrix

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
transposed_matrix= get_transpose(matrix)
print("Original:")
print(matrix)
print("Transposed:")
print(transposed_matrix)
```

### Problem 10: Flattening a Nested List

Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

```python
def flatten(nested_list):
    flattened_list = [item for row in nested_list for item in row]
    return flattened_list
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(f"The original matrix is : {matrix}")
print(f"The flat matrix is : {flatten(matrix)}")
```

### Problem 13: List Reversal

Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists.

```python
def reversed_list (numbers):
    reversed_numbers = numbers[::-1]
    return reversed_numbers
numbers = [random.randint(1,20) for i in range (10)]
print(f"The original list is : {numbers}")
print(f"The reversed list is : {reversed_list(numbers)}")
```

### Problem 14: List Rotation

Write a function that rotates a list by n positions. Print the original and rotated lists.

```python
def rotate_list(lst, n): 
    return lst[n:] + lst[:n]

original_list = [1, 2, 3, 4, 5]
rotated_list = rotate_list(original_list, 2)
print(f"Original list: {original_list}")
print(f"Rotated list: {rotated_list}")
```

### Problem 15: List Intersection

Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.

```python
def common_elements(first_list, second_list):
    instersected_lst = [i for i in first_list if i in second_list]
    return instersected_lst
lst1 = [1,2,3,4,5,6,7,8]
lst2 = [3,4,11,12,13,15]
print(common_elements(lst1,lst2))
```
