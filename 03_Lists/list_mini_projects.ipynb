##### Example 1. Manage A To Do List
- Create a To Do List To Keep Track OF Tasks

```python
Todo = []
while True:
    choice = input("Select the operations you want ? \nadd_task\nremove_task\nview_tasks\nexit ").lower()
    if choice == "exit":
        break
    elif choice == "add_task":
        task = input("Enter the new task : ").lower()
        if task in Todo:
            print("Task already exists please enter new task ")
        else :
            Todo.append(task)
    elif choice == "remove_task":
        task = input("Enter the task to be removed : ").lower()
        if task in Todo:
            print("Task removed successfully well done ")
            Todo.remove(task)
        else :
            print("Task doesnot exist please verify ")
    elif choice == "view_tasks":
        for task in Todo:
            print(task)
    else:
        print("wrong selection")
        
```

##### Example 2: Organizing Student Grades
- Create a list to store and calculate average grades for students

```python
marks = [int(mark) for mark in input("Enter the marks of all the students using spaces : ").split()]
print(marks)
average = sum(marks)/len(marks)
highest = max(marks)
lowest = min(marks)
print(
    f"Highest Score : {highest}\n"
    f"Lowest Score  : {lowest}\n"
    f"Average Score : {average:.2f}"
)
```

##### Example 3: Managing An Inventory
- Use a list to manage inventory items in a store

```python
inventory = []

while True:
    choice = input("Select the operation you want:\nadd_item\nremove_item\nview_inventory\nexit\n").lower()
    
    if choice == "exit":
        print("Exiting inventory management.")
        break
        
    elif choice == "add_item":
        item = input("Enter the new inventory item: ").lower()
        if item in inventory:
            print("Item already exists, please enter a new item.")
        else:
            inventory.append(item)
            print("Item added successfully.")
            
    elif choice == "remove_item":
        item = input("Enter the item to be removed: ").lower()
        if item in inventory:
            print("Item removed successfully, well done.")
            inventory.remove(item)
        else:
            print("Item does not exist, please verify.")
            
    elif choice == "view_inventory":
        if len(inventory) == 0:
            print("The inventory is currently empty.")
        else:
            for item in inventory:
                print(item)
        
    else:
        print("Wrong selection. Please try again.")
```

##### Example 4:Collecting User Feedback
- Use a list to collect and analyze user feedback.

```python
positive_terms = ["good","best","excellent","great","awesome"]
reviews = []
while True:
    positive_reviews = 0
    choice = input("Select the operation you want:\nadd_review\nview_review\nexit\nanalyze_review").lower()
    if choice == "exit":
        print("Exiting review system.")
        break
    elif choice == "add_review":
        review = input("Enter your review: ").lower()
        reviews.append(review)
        print("review added successfully.")
    elif choice == "view_review":
        if len(reviews) == 0:
            print("There is no review.")
        else:
            for review in reviews:
                print(review)
    elif choice == "analyze_review":
        if len(reviews) == 0:
            print("There is no review.")
        else:
            for review in reviews:
                for word in positive_terms:
                    if word in review:
                        positive_reviews+=1
                        break
            
        print(f"You have recievied {positive_reviews} numbers of positive review")
```
