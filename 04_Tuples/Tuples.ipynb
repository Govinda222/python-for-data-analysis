### Assignment 6: Tuple Methods

Create a tuple with duplicate elements and count the occurrences of an element. Find the index of the first occurrence of an element in the tuple.

```python
import random
duplicate_tup = tuple(random.randint(1,5) for i in range(15))
print(duplicate_tup)
print(f"The number of 1's in the tuple is {duplicate_tup.count(1)}")
print(f"The first occourance of 4 in the tuple is {duplicate_tup.index(4)}")
```

### Assignment 10: Tuple and List

### Assignment 14: Tuple and Set

Create a tuple with duplicate elements. Convert it to a set to remove duplicates and print the resulting set.

```python
import random
duplicate_tup = tuple(random.randint(1,5) for i in range(15))
print(duplicate_tup)
set_conversion = set(duplicate_tup)
print(set_conversion)
```

### Assignment 15: Tuple Functions

Write functions that take a tuple and return the minimum, maximum, and sum of the elements. Print the results for a sample tuple.

```python
def stats (tup):
    minimum = min (tup)
    maximum = max (tup)
    total = sum(tup)
    return(minimum,maximum,total)
tup7 = tuple(random.randint(1,100) for i in range (10))
print(tup7)
maximum,minimum,total = stats(tup7)
print(f"The sum is {total} the max is {maximum} and the minimum is {minimum}")
```
