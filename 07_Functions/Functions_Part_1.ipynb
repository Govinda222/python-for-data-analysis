##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.

```python
## why functions?
num=24
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
```

```python
## Call this function
even_or_odd(24)
```

```python
## function with multiple parameters

def add(a,b):
    return a+b

result=add(2,4)
print(result)
    
```

```python
## Default Parameters

def greet(name):
    print(f"Hello {name} Welcome To the paradise")

greet("Govinda")
```

```python
def greet(name="Guest"):
    print(f"Hello {name} Welcome To the paradise")

greet("Govinda")
```

```python
### Variable Length Arguments
## Positional And Keywords arguments

def print_numbers(*Govinda):
    for number in Govinda:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Govinda")
```

```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Govinda")
```

```python
### Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Govinda",age="23",country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional arument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Govinda",name="Govinda",age="23",country="India")
```

```python
### Return statements
def multiply(a,b):
    return a*b

multiply(2,3)
```

```python
### Return multiple parameters
def multiply(a,b):
    return a*b,a

multiply(2,3)
```
